```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

# Flatten true and predicted labels
y_true = []
y_pred = []

num_classes = conf_matrix.shape[0]

for pred in range(num_classes):
    for true in range(num_classes):
        count = int(conf_matrix[pred, true])  # Cast to int
        y_true.extend([true] * count)
        y_pred.extend([pred] * count)

# Accuracy
accuracy = accuracy_score(y_true, y_pred)

# Precision and Recall (macro-averaged over all classes)
precision = precision_score(y_true, y_pred, average='macro', zero_division=0)
recall = recall_score(y_true, y_pred, average='macro', zero_division=0)
precision_micro = precision_score(y_true, y_pred, average='micro', zero_division=0)
recall_micro = recall_score(y_true, y_pred, average='micro', zero_division=0)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision (macro): {precision:.4f}")
print(f"Recall (macro): {recall:.4f}")
print(f"Precision (micro): {precision_micro:.4f}")
print(f"Recall (micro): {recall_micro:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Class names
class_names = [
    'Biker', 'Car', 'Pedestrain', 'Trafficlight', 'Trafficlight-Green',
    'Trafficlight-Greenleft', 'Trafficlight-Red', 'Trafficlight-RedLeft',
    'Trafficlight-Yellow', 'Trafficlight-Yellowleft', 'Truck', 'Arret', 'background'
]

# Assuming conf_matrix is already defined and converted to float
conf_matrix = conf_matrix.astype(np.float64)

# True positives
tp = np.diag(conf_matrix)

# Precision
predicted_per_class = conf_matrix.sum(axis=1)
precision_per_class = np.divide(tp, predicted_per_class, out=np.zeros_like(tp), where=predicted_per_class!=0)

# Recall
actual_per_class = conf_matrix.sum(axis=0)
recall_per_class = np.divide(tp, actual_per_class, out=np.zeros_like(tp), where=actual_per_class!=0)

# F1 Score
f1_score_per_class = np.divide(
    2 * precision_per_class * recall_per_class,
    precision_per_class + recall_per_class,
    out=np.zeros_like(tp),
    where=(precision_per_class + recall_per_class) != 0
)

# Plotting
plt.figure(figsize=(14, 6))
bars = plt.bar(class_names, f1_score_per_class, color='mediumorchid', edgecolor='black')
plt.ylim(0, 1.05)
plt.title("Class-wise F1 Score")
plt.xlabel("Classes")
plt.ylabel("F1 Score")

# Annotate F1 score on top of bars
for bar, f1 in zip(bars, f1_score_per_class):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.01,
             f"{f1:.2f}", ha='center', va='bottom', fontsize=9)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
